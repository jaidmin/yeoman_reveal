```python
def step(input_list):
    return  [invert(elem) for elem in input_list]

def invert(x):
    if(x==1):
        return 0
    if(x==0):
        return 1
    
def sequence(length):
    seq = [0]
    for i in range(length):
        seq = seq + step(seq)
    return seq
```
